```python
def parser(x):
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

data = pd.read_csv('Print_Reading.csv',parse_dates =["pitime"], index_col ="pitime", low_memory=False, date_parser=parser)
data.head()
weekly_resample = data.resample('M')
weekly_resample
```
